```r
library(ggplot2)
influ<-read.csv(file.choose())

val <- c(influ$TotalA, influ$TotalB)
total <- data.frame(week=influ$Week , val, pp=influ$PercentPositive, ppA=influ$PercentPositiveA, ppB=influ$PercentPositiveB)
dist <-c(rep("TotalA",20),rep("TotalB",20))

#total
#df2 <- data.frame(week=influ$Week, pp=influ$PercentPositive)
#df2

#p <- ggplot(data=df2, aes(x=factor(df2$week), y=df2$pp,group = 1))+geom_line()

p <-ggplot(total)+
  
  geom_bar(aes(x = factor(total$week), y = val,fill = dist), stat="identity", color='black' )+
  theme(axis.text.x = element_text(angle=45, vjust=1, hjust=1))+xlab("Week") +
  ylab("Number of positive specimens")+scale_fill_manual(values=c("yellow2","darkgreen"),  name="",  labels=c("A", "B"))+
  scale_y_continuous(sec.axis = sec_axis(~./400, name = "Percent Positive"), breaks=seq(0,14000,2000))

  

#as.numeric(factor(df2$week))
#df2$pp

#p <- p + geom_line(aes(x=rep(as.numeric(factor(total$week)),2), y=rep((total$pp),2)*450))
bar_line <- p + geom_line(aes(x=(as.numeric(factor(total$week))), y=((total$pp))*400, color='1'), size=0.8, colour='black') +
            geom_line(aes(x=(as.numeric(factor(total$week))), y=((total$ppA))*400, color='2'), linetype = "dashed", size=0.8, colour='#E69F00') +
            geom_line(aes(x=(as.numeric(factor(total$week))), y=((total$ppB))*400, color='3'), linetype = "dashed", size=0.8, colour='green3')+  
labs(title = "\n\n  Influenza Positive Tests Reported to CDC by U.S. Clinical Laboratories, 
            2018-2019 Season\n\n")+
            scale_color_manual("",label=c("% Percent Positive" , "% Percent B","% Percent A"),
                               values=c("1"="black","2"="dark green","3"="yellow4"))
            #scale_fill_manual(name = "colors", values = c("yellow2","darkgreen", "black", "#E69F00" ,  "green3"), labels=c("A", "B","C","D","E"))
           
                 

bar_line 

#coord_polar("y")
```
